Faça um programa que verifique a 3ª lei de Kepler para os planetas do sistema  polar através de um gráfico de log(T2) x log(a3).
Iniciaremos o exercício extraindo oS dados da planilha fornecida e os colocando em um DataSet através da biblioteca pandas, para isso utilizaremos a função 
pandas.read_excel(), que lerá a planilha de dados anexada nos arquivos do Google Colab (equivalente dp Jupyter fornecido pelo Google).

```python
import pandas as pd
arquivo = '/AGA0521-SistemaSolar.xlsx'
pasta_SS = 'Sistema Solar'
df_SS = pd.read_excel(arquivo, sheet_name=pasta_SS, index_col=0)
```

A seguir realizaremos algumas operações através das colunas e as adicionaremos no DataSet. Essas operações consistem na criação de uma coluna com o valor do periodo orbital ao quadrado, T^2, o eixo da orbita ao cubo, a^3, log(T^2), log(a^3) e o valor de K=T^2/a^3.

```python
import numpy as np

t2_SS = np.array(df_SS['T'])**2  #em anos
df_SS['T^2'] = t2_SS  #em anos^2
a3_SS = np.array(df_SS['a'])**3 #em UA^3
df_SS['a^3'] = a3_SS  #em UA^3
logt2_SS = np.log(t2_SS)  #em log(anos^2)
df_SS['log(T^2)'] = logt2_SS #em log(anos^2)
loga3_SS = np.log(a3_SS) #em log(UA^3)
df_SS['log(a^3)'] = loga3_SS  #em log(UA^3)
k_SS = t2_SS/a3_SS  #em anos^2/UA^3
df_SS['K'] = k_SS  #em anos^2/UA^3
df_SS
```

Utilizando os dados da tabela acima, conseguimos plotar um gráfico que relaciona as colunas Log(a^3) e Log(T^2), nele podemos observar que existe uma relação linear entre as medidas. O que fáz sentido, já que a terceira Lei de Kepler nos diz que K=T^2/a^3 é uma contante. 
Portanto, o nosso gráfico abaixo plota uma reta que pode ser ajusta pela seguinte expressão: log(T^2) = log(K.a^3)


```python
import matplotlib.pyplot as plt

# Plot the points using matplotlib
plt.plot(loga3_SS, logt2_SS, marker = "*")
plt.xlabel('Log(a^3) [log(UA^3)]')
plt.ylabel(' Log(T^2) [log(anos^2)]')
plt.title('Verificação da 3ª lei de Kepler para os planetas do Sistema Solar')
plt.show()
```

  Uma segunda coisa que podemos observar, através da própria tabela seria que. de fato a variável K possuí uma constância, variando muito pouco de um planeta para outro. Com isso, podemos obter o valor médio e constatar que K = 1 [anos^2/UA^3].

```python
print( 'K_medio = ', df_SS['K'].mean(), 'anos^2/UA^3')
```

Modifique o programa da atividade anterior para verificar a 3ª lei de Kepler para os satélites Galileanos.

Realizaremos o mesmo processo feito anteriormente, onde observaremos a mesma relação linear com a diferença que para as Galileanas temos que K = 4.172597569993685e-17 [dias^2/Km^3].

```python
pasta_G = 'Galileanas'
df_G = pd.read_excel(arquivo, sheet_name=pasta_G, index_col=0)

t2_G = np.array(df_G['T'])**2 #em dias^2
df_G['T^2'] = t2_G #em dias^2
a3_G = np.array(df_G['a'])**3 #em km^3
df_G['a^3'] = a3_G #em km^3
logt2_G = np.log(t2_G) #em log(dias^2)
df_G['log(T^2)'] = logt2_G #em log(dias^2)
loga3_G = np.log(a3_G) #em log(km^3)
df_G['log(a^3)'] = loga3_G #em log(km^3)
k_G = t2_G/a3_G #em dias^2/km^3
df_G['K'] = k_G  #em dias^2/km^3
df_G
```

```python
# Plot the points using matplotlib
plt.plot(loga3_G, logt2_G, marker = "*")
plt.xlabel('Log(a^3) [log(Km^3)]')
plt.ylabel(' Log(T^2) [log(dias^2)]')
plt.title('Verificação da 3ª lei de Kepler para as Galileanas')
plt.show()
```
