Os motores dos mísseis V2 (vide Aula 8), utilizados no final da II Guerra Mundial, ofereciam um impulso específico de 215 s com um empuxo de 242307 N. Desenvolva um programa que calcule:

a) A velocidade de escape dos gases


Apresente todos os resultados intermediários com as unidades adequadas.

Para o calculo da velocidade de escape dos gases, c, utilizarei a expressão: c = I_sg*g_0, onde g_0 é a gravidade da terra

```python
#definindo os parâmetros
g_0 = 9.8 #em m/s^2
I_sg = 215 #em segundos 

c = g_0*I_sg #em m/s

print('c =', c, 'm/s')
```

b) A taxa mássica (kg/s)

Para obtermos o valor da taxa mássica, utilizaremos a expressão: dm_e = T/c, onde c é a velocidade de escape dos gases e T é o empuxo

```python
#definindo os parâmetros
T = 242307 #em N
dm_e = T/c

print('dm_e =', dm_e, 'kg/s')
```
