1. Qual é a massa de Beta Tauri? Quantos anos tem o período orbital de
Efepirus?

Como sistema Beta Tauri possui dois planetas, podemos considerar que a massa do mesmo pode ser aproximada pela soma dos planetas que o compões sendo eles: Taurus II e Efepirus. Dito isso temos  

```python
massa_BetaTauri = m_efepirus + m_taurus #em kg

print('Massa de Beta Tauri = ',massa_BetaTauri, 'kg')
```

Para calcular o periodo orbital de Efepirus, utilizaremos a expressão a seguir para obter o valor: T_efepirus = 2*pi*(a_eferius^(3/2))/(u_efepirus^(1/2)), onde u_eferius = G*m_eferius


```python
import numpy as np

G = 6.674e-20 #em N(km/kg)^2

T_efepirus_segundo =  2*np.pi*(a_efepirus**3/2)/(G*m_efepirus**(1/2)) #em segundos
T_efepirus_horas = T_efepirus_segundo/3600 #em horas

anos_dias = 365.12 #em dias
dias_horas = 24#em horas

T_efepirus_anos = T_efepirus_horas/(dias_horas*anos_dias) #em anos

print('T_eferius = ',T_efepirus_anos, 'anos')

T_taurus_anos = T_taurus/(dias_horas*anos_dias) #em anos

print('T_taurus = ',T_taurus_anos, 'anos')
```

2. Calcule o Dv necessário para a Enterprise alcançar uma órbita de
transferência para Efepirus a partir da órbita 2 e o ângulo do perigeu da
hipérbole de partida em graus.


Inicialmente, a Enterprise estava em uma órbita elíptica ao redor de Taurus II, de 300 km por 6420 km (órbita 1).

A seguir registraremos também esse parâmetros:

```python
#parametros da orbita 1 em Taurus

#altura mínima órbita 1
z_a1 = 300 #em km
#altura maxima órbita 1
z_c1 = 6420 #em km

#orbita 1 mínima 
r_a1 = r_taurus + z_a1 #em km
#orbita 1 maxima
r_c1 = r_taurus + z_c1 #em km

#Calculando a excentricidade da órbita 1, que é um parâmetro adimensional
e1 = (r_c1 - r_a1 )/(r_c1 + r_a1 )
print("e1 = ", e1)
```

Sabendo que a equação orbital no perigeu é dada pela expressão: r_a = (h1^2/u)*(1/(1+e1*cos(0)) -> h1 = (r_a*u*(1+e1))^(1/2), como u = G*(m1 + m1) e a massa de Taurus é muito maior que a da espaçonave, consideramos u = G*m1.
Portanto, h1 = (r_a*G*M_terra*(1+e1))^(1/2)

```python
h1 = np.sqrt((r_a1)*G*m_taurus*(1+e1)) #km^2/ s, 
print("h1 = ", h1 , 'km^2/ s')
```

Para obter a velocidade no perigeu da órbita 1, farei va_1 = h1/r_a1

```python
va_1 = h1/r_a1 # em km/s
print('va_1 = ', va_1, "km/s")
```


O plano do capitão Kirk era iniciar a  viagem interplanetária a partir de uma órbita circular de 11500 km acima da superfície do planeta (órbita 2).

```python
#altura mínima órbita 2
z_a2 = 11500 #em km
#altura maxima órbita 2
z_c2 = 11500 #em km

#orbita 2 mínima 
r_a2 = r_taurus + z_a2 #em km
#orbita 2 maxima
r_c2 = r_taurus + z_c2 #em km

#Calculando a excentricidade da órbita 2, que é um parâmetro adimensional
e2 = (r_c2 - r_a2 )/(r_c2 + r_a2 )
print("e2 = ", e2)
```

```python
h2 = np.sqrt((r_a2)*G*m_taurus*(1+e2)) #km^2/ s, 
print("h2 = ", h2 , 'km^2/ s')
```

```python
va_2 = h2/r_a2 # em km/s
print('va_2 = ', va_2, "km/s")
```

```python
#diferença de velocidade entre as órbitas 1 e 2
DeltaVa = va_2 - va_1 # em km/s
print('Delta_va= ', DeltaVa, "km/s")
```

Para calcular o ângulo do perigeu da
hipérbole de partida em graus, iniciaremos os calculos utilizando a expressão v_infinito = ((u_BetaTauri/a_taurus)^(1/2))(((2a_eferipus/(a_taurus+a_eferipus))^(1/2)) - 1) para calcular a velocidade de excesso hiperbólico

```python
v_infinito = np.sqrt((G*massa_BetaTauri)/a_taurus)*(np.sqrt(2*a_efepirus/(a_taurus+a_efepirus)) - 1) #em km/s
print('v_infinito = ', v_infinito,'km/s')
```

Já para calcular a velocidade da nave na órbita circular inicial usamos v_c = (u_taurus/(r_taurus + z_orbita))^(1/2), onde z_orbita é a altura da orbita

```python
v_c = np.sqrt((G*m_taurus)/(r_taurus + z_a2)) #em km/s

print('v_c = ', v_c,'km/s')
```

Para o perigeu da hipérbole de partida, em relação o vetor de velocidade orbital da Efepirus, utilizamos a expressão b = arccos((1 + (r_p*(v_infinito^2)/u_taurus) )^(-1)), onde r_p é a soma do raio terrestre com a altura da orbita

```python
r_p = r_taurus + z_a2
b_rad = np.arccos((1 + (r_p*(v_infinito**2)/(G*m_taurus)) )**(-1)) #em radianos
b_g = b_rad*180/np.pi
print('b_g = ', b_g, '°')
```

3. Quantos dias tem o período sinódico entre Taurus II e Efepirus?

Para calcular o período sinódico utilizaremos a expressão: Tsyn  = (T_taurus*T_efepirus)*(|T_taurus - 
T_efepirus|).


```python
Tsyn = (T_taurus_anos*T_efepirus_anos)/(np.abs(T_taurus_anos - T_efepirus_anos)) #em anos
print('Tsyn = ',Tsyn*365.12, 'dias')
```

```python
T_orbital = np.pi*(((1/2)*(r_b2+r_a2))**(3/2))/((G*m_taurus)**(1/2)) # em segundos
print('Tempo de transferência orbital = ',T_orbital/3600, 'horas')
```
