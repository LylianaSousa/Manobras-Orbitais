Desenvolva um programa simples que calcula o módulo da força  gravitacional entre dois corpos, dados:

• A massa de cada corpo

• A distância entre eles

Verifique a validade do programa com os exemplos 1 e 2 da aula sobre
as Leis de Newton.

Para a criação desse programa utilizaremos a Lei Universal da Gravitação desenvolvida por Newton, que nos diz que: F_grav = G*m_1*m_2/r^2, onde consideraremos a constante gravitacional como sendo G = 6.674*10^(-11) [N(m/Kg)^2].

```python
11#Definindo as variáveis
m_1 = float(input('Digite a massa do primeiro corpo, com as unidades em Kg: '))
m_2 = float(input('Digite a massa do segundo corpo, com as unidades em Kg: '))
r = float(input('Digite a distância entre os corpos, em metros: '))
G = 6.674*(10**(-11) )

#Definindo o modulo da força gravitacional
F_grav = G*m_1*m_2/(r**2)
print('O modulo da força gravitacional é F_grav = {0} [N].'.format(F_grav))
```

Desenvolva um programa que calcula a massa do corpo central de uma órbita e teste-o com o exemplo do Sol.

Para obtermos a relação com a massa do corpo central, utilizaremos a Lei universal da Gravitação e a Segunda lei de Newton. Desta maneira, poderemos considerar que o modulo da força gravitacional é igual ao modulo da força centripeta. Portanto, temos a seguinte relação:
F_grav = G*m_1*m_2/r^2 = m_1*v^2/r = 4*pi*m_1/T^2 -> m_2 = 4*(pi^2)*(r^3)/G*(T^2)


```python
import numpy as np

T = float(input('Digite o periodo orbital, em segundos: '))
r = float(input('Digite a distância entre os corpos, em metros: '))
#Definindo o modulo da força gravitacional
M = 4*(np.pi**2)*(r**3)/(G*(T**2))
print('A massa do objeto será dada por: ', f'{M:.5e}','[kg]')
```
