Elabore um programa que resolva o exemplo da transferência de Hohmann simples apresentado na Aula 3, explicitando todos os resultados intermediários e o resultado final.
Uma espaçonave está em uma órbita elíptica em torno
da Terra de 480 km por 800 km (órbita 1). 


```python
#Definindo os parâmetros
r_terra = 6378 #em km

#altura mínima órbita 1
z_a1 = 480 #em km
#altura maxima órbita 1
z_c1 = 800 #em km

#orbita 1 mínima 
r_a1 = r_terra + z_a1 #em km
#orbita 1 maxima
r_c1 = r_terra + z_c1 #em km

#Calculando a excentricidade da órbita 1, que é um parâmetro adimensional
e1 = (r_c1 - r_a1 )/(r_c1 + r_a1 )
print("e1 = ", e1)
```

Sabendo que a equação orbital no perigeu é dada pela expressão: r_a = (h1^2/u)*(1/(1+e1*cos(0)) -> h1 = (r_a*u*(1+e1))^(1/2), como u = G*(m1 + m1) e a massa da terra é muito maior que a da espaçonave, consideramos u = G*m1.
Portanto, h1 = (r_a*G*M_terra*(1+e1))^(1/2)

```python
import numpy as np

G = 6.674e-20 #em N(km/kg)^2
M_terra = 5.98e24 #em kg
h1 = np.sqrt((r_a1)*G*M_terra*(1+e1)) #km^2/ s, 
print("h1 = ", h1 , 'km^2/ s')
```

Para obter a velocidade no perigeu da órbita 1, farei va_1 = h1/r_a1

```python
va_1 = h1/r_a1 # em km/s
print('va_1 = ', va_1, "km/s")



```

(a) Encontre o Deltav necessário no perigeu (A) para colocar a sondab numa órbita de transferência (órbita 2) de 480 km por 16000 km.



```python
#altura mínima órbita 2
z_a2 = 480 #em km
#altura maxima órbita 2
z_c2 = 16000 #em km

#orbita 2 mínima 
r_a2 = r_terra + z_a2 #em km
#orbita 2 maxima
r_c2 = r_terra + z_c2 #em km

#Calculando a excentricidade da órbita 2, que é um parâmetro adimensional
e2 = (r_c2 - r_a2 )/(r_c2 + r_a2 )
print("e2 = ", e2)
```

```python
h2 = np.sqrt((r_a2)*G*M_terra*(1+e2)) #km^2/ s, 
print("h2 = ", h2 , 'km^2/ s')
```

```python
va_2 = h2/r_a2 # em km/s
print('va_2 = ', va_2, "km/s")
```

```python
#diferença de velocidade entre as órbitas 1 e 2
DeltaVa = va_2 - va_1 # em km/s
print('Delta_va= ', DeltaVa, "km/s")
```

```python
#altura maxima órbita 3
z_c3 = 16000 #em km

#orbita 2 mínima 
r_b2 = r_terra + z_c3 #em km

vb_2 = h2/r_b2 # em km/s
print('vb_2 = ', vb_2, "km/s")
```

Devido a órbita circular, sua velocidade é expressada pela relação: v_circular = (u/r)^(1/2) = (G*R_terra/rb_2)^(1/2)

```python
vb_3 = np.sqrt(G*M_terra/r_b2)
print('v_circular = ',vb_3, "km/s")
```

(b) Encontre o Deltav no apogeu (B) necessário na órbita de transferência B para estabelecer uma órbita circular de 16000 km de altitude (órbita 3).

```python
#diferença de velocidade entre as órbitas 2 e 3 (no apogeu)
DeltaVb = vb_3 - vb_2
print('Delta_vb = ', DeltaVb, "km/s")
```

```python
DeltaVtotal = np.abs(DeltaVa) + np.abs(DeltaVb)
print('DeltaVtotal = ', DeltaVtotal, "km/s")
```

A tempo da  transferência orbital é dado por metade do periodo, T_orbital = pi*(a^(3/2))/(u^(1/2))

```python
T_orbital = np.pi*(((1/2)*(r_b2+r_a2))**(3/2))/((G*M_terra)**(1/2)) # em horas
print('Tempo de transferência orbital = ',T_orbital/3600, 'horas')
```
