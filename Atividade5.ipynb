Uma nave espacial é lançada em uma missão para Marte a partir de uma órbita circular inicial de 300 km.

Iniciaremos a atividade definindo os parâmetros utilizados para os calculos do DeltaV, da localização do perigeu da hipérbole partida e da quantidade de propelente requerida como uma percentagem da massa da sonda antes da queima do Dv, assumindo um impulso específico de 300 segundos.

```python
#Os parâmetros a seguir foram obtidos da tabela de slides do enunciado
u_sol = 1.327e11 #em km^3/s^2
u_terra = 398600 #em km^3/s^2
R_terra = 149.6e6 #em km (semieixo)
R_marte = 227.9e6 #em km (semieixo)
r_terra = 6378 #em km
r_marte = 3396 #em km
```

  Iniciaremos os calculos utilizando a expressão v_infinito = ((u_sol/R_terra)^(1/2))*(((2*R_marte/(R_terra+R_marte))^(1/2)) - 1) para calcular a velocidade de excesso hiperbólico

```python
import numpy as np
v_infinito = np.sqrt(u_sol/R_terra)*(np.sqrt(2*R_marte/(R_terra+R_marte)) - 1) #em km/s
print('v_infinito = ', v_infinito,'km/s')
```

Já para calcular a velocidade da nave na órbita circular inicial usamos v_c = (u_terra/(r_terra + z_orbita))^(1/2), onde z_orbita é a altura da orbita

```python
z_orbita = 300 #em km
v_c = np.sqrt(u_terra/(r_terra + z_orbita)) #em km/s

print('v_c = ', v_c,'km/s')
```

Por fim, para obtermos a variação da velocidades utilizaremos a seguinte expressão:


.

 DeltaV = v_c*(((2 + ((v_infinito/v_c)^2)^(1/2)) - 1)

```python
DeltaV = v_c*(np.sqrt(2 + ((v_infinito/v_c)**2)) - 1) #em km/s
print('DeltaV = ', DeltaV,'km/s')
```

Para o perigeu da hipérbole de partida, em relação  o vetor de velocidade orbital da Terra, utilizamos a expressão b = arccos((1 + (r_p*(v_infinito^2)/u_terra) )^(-1)), onde r_p é a soma do raio terrestre com a altura da orbita

```python
r_p = r_terra + z_orbita
b_rad = np.arccos((1 + (r_p*(v_infinito**2)/u_terra) )**(-1)) #em radianos
b_g = b_rad*180/np.pi
print('b_g = ', b_g, '°')
```

Por fim, finalizaremos a atividade obtendo a razão entre a diferença de massa e o valor da massa inicial, a partir da relação a seguir: Deltam_minicial = 1 - e^(-Deltav/I_sp*g_0), onde I_sp sendo o impulso e g_0 a gravidade.

```python
g_0 = 9.81e-3 #em km/s^2
I_sp = 300 #em s
Deltam_m0 = 1 - np.exp((-1*DeltaV)/(I_sp*g_0)) #resultado adimensional
print('Deltam_m0 = ', Deltam_m0)
```

Calcule o Dv mínimo necessário para colocar uma nave espacial em órbita de Marte com um período de sete
horas, sabendo-se que a manobra de chegada foi uma transferência de Hohmann.
O semieixo maior da órbita de captura em função do período é dado por: a  = (T*(u_marte^(1/2))/2*pi)^(2/3)

```python
T_horas = 7 #em horas
T_segundos = T_horas*3600 #em segundos
u_marte = 42830 # em km^3/s^2 
a  = (T_segundos*(u_marte**(1/2))/(2*np.pi))**(2/3)
print('a = ', a , 'km')
```

```python
v_infinito_marte = np.sqrt(u_sol/R_marte)*np.abs(np.sqrt(2*R_terra/(R_terra+R_marte)) - 1) #em km/s
print('v_infinito_marte = ', v_infinito_marte,'km/s')
```

Sabendo que a excentricidade é calculada por e = ((2*u_marte)/(a*v_infinito^2)) - 1

```python
e = ((2*u_marte)/(a*(v_infinito_marte**2))) - 1 #adimensional
print('e = ', e )
```

Por fim, o deltav_marte será dado pela expressão: deltav_marte = v_infinito_marte*(((1-e)/2)^(1/2))

```python
deltav_marte = v_infinito_marte*np.sqrt((1-e)/2)
print('deltav_marte = ', deltav_marte,'km/s')
```

Para calcular o período sinódico utilizaremos a expressão: Tsyn  = (T_terra*T_marte)*(|T_terra - 
T_marte|)

```python
T_terra = 365.26 #em dias
T_marte = 687.05 #em dias
Tsyn  = (T_terra*T_marte)/(np.abs(T_terra - T_marte)) #em dias
print('Tsyn = ',Tsyn, 'dias')
```
