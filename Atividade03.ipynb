 Elabore um programa que utilize a posição de um corpo em órbita da Terra para determinar sua ascensão reta (a) e declinação (d).

 Para isso, iniciaremos definindo o vetor posição, onde o algoritmo formará um array onde r = [rx,ry,rz].
 Definiremos abaixo o vetor posição que usaremos ao longo do programa. Nesse caso, utilizaremos o exemplo da Estação Espacial Internacional
apresentado na Aula 2.


```python
import numpy as np

#Definindo o vetor posição
 
#posição da ISS – Estação Espacial Internacional
rx = -5368 #km
ry = -1784 #km
rz = 3691 #km

#Modulo de r valerá
r = np.sqrt((rx**2) +(ry**2) +(rz**2) ) 
print(r, 'km')
```

Lembrando que o vetor unitário de um referenciaç equatorial geocêntrico é definido como u_r = [u_rx, u_ry, u_rz] = [cos(d)*cos(a), cos(d)*sen(a), sen(d)]

```python
#Obtendo o vetor unitário
u_rx = rx/r
u_ry = ry/r
u_rz = rz/r

print("u_rx = ", u_rx, 'km')
print("u_ry = ", u_ry, 'km')
print("u_rz = ", u_rz, 'km')
```

Sandendo que u_rz = sen(d) por definição, temos que o ângulo de declição pode ser obtido a partir do arco-sen de u_rz: d = arcsen(u_rz).

```python
#Obtendo o ângulo de declinação


d_rad = np.arcsin(u_rz) #em radianos
d_g = d_rad*180/np.pi #em graus
print("d = ", d_g,"°")
```

Para determinarmos o valor da ascensão da reta, podemos utilizar a relação a seguir para obter os valores de seno e cosseno da ascensão.
Cosa = u_rx/cos(d) e Sena = u_ry/cos(d), onde para obtermos o angulo correto da ascensão devemos realizar uma transformação, pós calculo da arco-seno do seno, levando em considerção o quadrante que estamos e o final que a função arcseno devolve.

```python
#Obtendo os valores de cosseno e seno da ascensão da reta

cosa = u_rx/np.cos(d_rad)
sena = u_ry/np.cos(d_rad)
print("cosa = ", cosa,"rad")
print("sena = ", sena,"rad")


#Construindo uma relação que converterá os resultados a partir do quandrante em que o vetor a se encontra
if cosa >= 0 and sena >= 0:
  print("Estamos no quadrante 1")
  a_rad = np.arcsin(sena) #em radianos
elif cosa <= 0 and sena >= 0:
  print("Estamos no quadrante 2")
  a_rad = (np.pi) - np.arcsin(sena) #em radianos
elif cosa <= 0 and sena <= 0:
  print("Estamos no quadrante 3")
  a_rad = (np.pi) - np.arcsin(sena) #em radianos
elif cosa >= 0 and sena <= 0:
  print("Estamos no quadrante 4")
  a_rad = (2*np.pi) - np.arcsin(sena) #em radianos

#convertendo o valor da ascensão da reta para graus
a_g = a_rad*180/np.pi #em graus
print("a = ", a_g,"°")
```
