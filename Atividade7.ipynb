- Massa inicial = 2030 toneladas
- Massa de combustível consumida = 1272000 kg

Dica: use a equação do foguete, e, não se esqueça de explicitar todos os

Na aula 3, nos fomos apresentados há uma equação que relacionava a variação da massa, DeltaM, e a massa inicial, m_0, com a variação da velocidade, DeltaV, com o impulso específico, I_sg,  e a gravidade que atua sob o corpo, g_0: (DeltaM/m_0) = 1 - exp(-DeltaV/(I_sg*g_0)). A partir dessa expressão, podemos manipulará para obter uma relação onde dada o DeltaM, DeltaV, m_0 e o g_0, podemos encontrar o I_sg. Após as manipulações, obtemos a seguinte relação: I_sg = - (DeltaV/g_0)*((ln(1 - (DeltaM/m_0)))^(-1))

```python
import numpy as np

#definindo os parâmetros
DeltaM =  1272000 #em kg
DeltaV_km = 3.0522 #em km/s
DeltaV_m = DeltaV_km*1000 #em m/s
m_0 = 2030000 #em kg
g_0 = 9.8 #em m/s^2

I_sg = - (DeltaV_m/g_0)*((np.log(1 - (DeltaM/m_0)))**(-1)) #em segundos

print('I_sg = ', I_sg, 's')
```
